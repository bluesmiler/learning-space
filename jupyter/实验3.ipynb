```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set(style='darkgrid')
plt.rcParams['font.sans-serif'] = 'SimHei'  
plt.rcParams['axes.unicode_minus'] = False
data = pd.read_csv('./aqi.csv')
ax=sns.scatterplot(x='AQI',y='PM2.5含量（ppm）',data=data)
plt.show()
```

```python
sns.stripplot(x='质量等级',data=data)
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
sns.displot(x="质量等级",data=data)
plt.title('单变量的分布图')
plt.ylabel('数量')
plt.show()
```

```python
sns.regplot(x='PM2.5含量（ppm）', y='AQI', data=data, ci=50)
plt.show()
```

```python
plt.rcParams['axes.unicode_minus'] = False
corr = data.corr()  
sns.heatmap(corr)
plt.title('特征矩阵热力图')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(corr, annot=True, fmt='.2f')
plt.title('特征矩阵热力图')
plt.show()
```
