```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = 'SimHei'  
plt.rcParams['axes.unicode_minus'] = False
data = pd.read_csv('./employee.csv')
name = data.columns  
values = data.values  
plt.figure(figsize = (8,7))   
plt.scatter(values[:,0],values[:,2], marker = 'o',c = 'red')  
plt.scatter(values[:,0],values[:,3], marker = 'D',c = 'blue')  
plt.scatter(values[:,0],values[:,4], marker = 'v',c = 'black')  
plt.xlabel('')  
plt.ylabel('')  
plt.xticks(range(0,20),values[range(0,20),0],rotation = 45)
plt.title('2000-2019')  
plt.legend(['','',''])   
plt.show()
```

```python
plt.figure(figsize = (7,6)) 
plt.plot(values[:,0],values[:,2],'bs-',
       values[:,0],values[:,3],'ro-.',
       values[:,0],values[:,4],'gH--')  
plt.xlabel('')  
plt.ylabel('')  
plt.xticks(range(0,20),values[range(0,20),0],rotation = 45)
plt.title('2000-2019') 
plt.legend(['','',''])  
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = 'SimHei'  
plt.rcParams['axes.unicode_minus'] = False
data = pd.read_csv('./employee.csv')
name = data.columns  
values = data.values  
label = ['','','']  
plt.figure(figsize = (4,4))  
explode = [0.01,0.01,0.01] 
plt.pie(values[-1,2:5],explode = explode,labels = label,autopct = '%1.1f%%')  #
plt.title('2019')  
plt.show()
```

```python
label = ['','','']  
plt.figure(figsize = (9,8), dpi = 60)
plt.bar(range(3),values[-1, 2: 5],width = 0.3)
plt.xlabel('') 
plt.ylabel('')  
plt.xticks(range(3),label)
plt.title('2019')  
plt.show()
```

```python
label= ['','','']  
gdp = (list(values[:,2]),list(values[:,3]),list(values[:,4]))
plt.figure(figsize=(6,5))
plt.boxplot(gdp,notch = True,labels = label,showmeans=True, meanline = False)
plt.title('2000-2019')
plt.show()
```
