```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
data1=pd.read_csv('./数据1.csv',encoding='gbk',header='infer')
data2=pd.read_csv('./数据2.csv',encoding='gbk',header='infer')
print(data1)
print(data2)
```

```python
data=pd.merge(data1,data2,on='商品')
print(data)
```

```python
data.T
```

```python
data.to_excel('./data.xlsx',encoding='gbk')
```

```python
print('before: ',data.shape)
data.drop_duplicates()
print('after:',data.shape)
```

```python
print('before: ',data.shape)
data.drop_duplicates()
print('after:',data.dropna(axis=1,how='any').shape)
```
